```python
#even odd
def even_odd(num):
  if(num % 2 == 0):
    return "even"
  else:
    return 'Odd'

print(even_odd(10))
```

```python
#Reverse of a number
def reverseNumber(num):
  temp = num
  reverse = 0
  while(num > 0):
    remainder = num % 10
    reverse = (reverse * 10) + remainder
    num = num // 10
  return reverse

print(reverseNumber(123456))
```

```python
#String palindrome
def stringPal(str):
  if str == str[::-1]: #slicing
    return 'Palindrome'
  return 'Not palindrme'

print(stringPal('racecar'))
print(stringPal('madam'))
print(stringPal('hello'))
```

```python
#Chek leap year
def checkLeapYear(year):
  if((year % 400 == 0) or (year%4==0 and year%100!=0)): return f'{year} is a leap year' #ekkada f antey f-strings
  return f'{year} is not a leap year'

print(checkLeapYear(2001))
print(checkLeapYear(2020))
```

```python
#Sum of first n natural numbers
def sumNaturals(num):
  sum = 0
  for i in range(num):
    sum += i
  return sum

print(sumNaturals(8))
print(sumNaturals(13))

```

```python
# Number Palindrome

def numPalindrome(num):
  reverse = 0
  temp = num
  while temp > 0:
    remainder = temp % 10
    reverse = (reverse * 10) + remainder
    temp = temp // 10

  if(num == reverse): return f'{num} is a palindrome number'
  return f'{num} is not a palindrome number'

print(numPalindrome(143))
print(numPalindrome(113311))


```

```python
# Power of a number

def powerOfNum(num, power):
  result = pow(num, power)
  return result

print(powerOfNum(5, 2))
print(powerOfNum(2, 2))
```
